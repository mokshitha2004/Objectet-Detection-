```python
import os
os.listdir("cat_dataset")
```

```python
from ultralytics import YOLO

# Load pre-trained YOLOv8 model (nano version)
model = YOLO("yolov8n.pt")

# Start training
model.train(
    data="cat_dataset/data.yaml",
    epochs=30,
    imgsz=640,
    batch=16
)
```

```python
metrics = model.val()
```

```python
results = model.predict(source="cat_dataset/test/images", save=True)
```
